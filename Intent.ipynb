```python
import re
import csv
import emoji

def replace_short_forms(sentence):
    short_forms = {}
    with open('acronym.csv', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for row in reader:
            short_forms[row[0]] = row[1]
    
    words = sentence.split()
    for i, word in enumerate(words):
        if word.lower() in short_forms:
            words[i] = short_forms[word.lower()]
    
    return ' '.join(words)

def preprocess(text):
    # Convert to lowercase
    text = text.lower()
    
    # Convert acronym to actual
    text = replace_short_forms(text)
    
    # Convert emoji to text
    text = emoji.demojize(text)
    text = text.replace(":", "")
    text = replace_emoji_labels(text)
    text = remove_repeat(text)
    
    return text

text = "I Want to Set up a meeting"
processed_text = preprocess(text)
print(processed_text)
```

```python
lab = LabelEncoder()
intents = df_train['intent']
intents_transform = lab.fit_transform(intents)
```

```python
Classification_Model = make_pipeline(StandardScaler(), OneVsRestClassifier(SVC(kernel="poly", C=200, gamma='auto', probability=True)))
Classification_Model.fit(embeddings,intents_transform)
```

```python
# Function to predict class and probability score for a single input sentence
import numpy as np
def predict(sentence):
    embeddings = modelroberta.encode([sentence])
    proba = Classification_Model.predict_proba(embeddings)
    pred_idx = np.argmax(proba)
    classes = lab.classes_
    return classes[pred_idx], proba[0][pred_idx]
```

```python
# Example usage
sentence = "Yesterday"
print("Input Sentence:", sentence)
#sentence = preprocess(sentence)
print("Preprocessed Sentence:", sentence)
intent, proba = predict(sentence)


print("Predicted Intent:", intent)
print("Probability Score:", proba)
```
